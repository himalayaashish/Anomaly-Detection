```python
import statsmodels.api as sm
# Resample or fill missing minute intervals to ensure regular frequency
df = df.resample('min').mean()  # 'T' for minute frequency

# Check for missing values and fill them if necessary
df['Readings'] = df['Readings'].ffill()#, inplace=True)  # Forward-fill missing values
# df.method({'Readings': 'ffill'}, inplace=True)

# Set the period explicitly (1440 for one day of minute-level data)
period = 1440

# Decompose the time series using seasonal_decompose (additive model)
model = sm.tsa.seasonal_decompose(df['Readings'], model='additive', period=period)

model.plot();
# import matplotlib.pyplot as plt
# plt.show()
```

```python
from sklearn.metrics import (
    mean_squared_error as mse,
    mean_absolute_error as mae,
    mean_absolute_percentage_error as mape
)

# Creating a function to print values of all these metrics.
def performance(actual, predicted):
    print('MAE :', round(mae(actual, predicted), 3))
    print('RMSE :', round(mse(actual, predicted)**0.5, 3))
    print('MAPE:', round(mape(actual, predicted), 3))
```

```python

train_x = df.loc[df.index < df.index[-5000]].copy()
# df.loc[df.index <= '02-01-2017']
test_x = df.loc[df.index >= df.index[-5000]].copy()
```

##### Seasonal Naive

```python
for i in test_x.index:
#     print(i)
#     test_x.loc[i]['pred'] = train_x.loc[i - pd.DateOffset(days=5)].Readings
    test_x.at[i, 'pred'] = train_x.loc[i - pd.DateOffset(days=10), 'Readings']
#     print(test_x['pred'])




test_x.plot(style='-o')

performance(test_x['Readings'], test_x['pred'])
```

##### Drift forecasting

```python
y_t = train_x['Readings'][-1]
y_0 = train_x['Readings'][0]

m = (y_t - y_0) / len(train_x['Readings'])

h = np.arange(0, len(test_x))
```

```python
# h*m
```

```python
test_x['pred'] = y_t + h*m
```

```python

test_x.plot(style='-o')

performance(test_x['Readings'], test_x['pred'])
```

```python
def adf_test(data, significance_level=0.05):
    pvalue = sm.tsa.stattools.adfuller(data)[1]

    if pvalue <= significance_level:
        print('Sequence is stationary')
    else:
        print('Sequence is not stationary')
```

```python
adf_test(df.Readings)
```
